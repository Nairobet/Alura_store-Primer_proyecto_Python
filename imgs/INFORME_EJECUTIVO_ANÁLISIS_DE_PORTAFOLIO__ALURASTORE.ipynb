# INFORME EJECUTIVO: ANÁLISIS DE PORTAFOLIO ALURASTORE

**Fecha:** Diciembre 2025  
**Período analizado:** 3 años de operaciones (4 tiendas físicas)  
**Objetivo:** Evaluar desempeño de cada tienda para optimizar portafolio

---

## 1. SITUACIÓN ACTUAL: FACTURACIÓN POR TIENDA

### Resumen financiero consolidado

En los últimos 3 años, AluraStore ha acumulado una facturación total de **4.403.619.200 de pesos** distribuida entre sus 4 tiendas.

| Tienda | Facturación | % del Total | Posición |
|--------|------------|-----------|----------|
| Tienda 1 | 1.150.880.400 | 26,1% |  1era |
| Tienda 2 | 1.116.343.500 | 25,3% |  2da |
| Tienda 3 | 1.098.019.600 | 24,9% |  3era |
| Tienda 4 | 1.038.375.700 | **23,6%** | 4ta |

**Hallazgo clave:** Las 3 primeras tiendas mantienen un desempeño muy similar (25–26% cada una), mientras que **la Tienda 4 rezaga 2–3 puntos porcentuales**, acumulando el 23,6% de la facturación con menor participación relativa.

---

## 2. ANÁLISIS DE PRODUCTOS: CONCENTRACIÓN DE VENTAS

El **Top 5 de productos** concentra el **32.3%** de toda la facturación, indicando una cartera muy enfocada. Esto revela:

**Fortaleza:** Líneas de productos claras y demanda predecible.  
**Riesgo:** Dependencia de pocos SKUs. Una interrupción afecta mucho.

**Recomendación operativa:** Asegurar stock continuo de los 5 top productos y explorar complementarios en categorías de menor venta.

---
## 3. SATISFACCIÓN DEL CLIENTE: CALIFICACIÓN PROMEDIO (CORREGIDO)

Las cuatro tiendas mantienen un nivel de satisfacción **alto y homogéneo**, con calificaciones promedio entre **3,98 y 4,05** sobre 5. Esto indica que la experiencia del cliente es una fortaleza transversal del negocio.

| Tienda   | Calificación promedio |
|----------|-----------------------|
| Tienda 3 | **4,05** |
| Tienda 2 | **4,04** |
| Tienda 4 | **4,00** |
| Tienda 1 | **3,98** |

**Conclusión:** Tienda 4 no es la peor en satisfacción; su experiencia de cliente es comparable al resto. La decisión sobre su continuidad debe basarse principalmente en indicadores financieros y de escala cómo facturación, participación y volumen, no en servicio.

**Hallazgo:** Tienda 4 cubre ciudades similares a Tienda 1, 2 y 3, pero con menor penetración de mercado en cada una.

## 7. RECOMENDACIÓN ESTRATÉGICA

Basado en el análisis integral, recomendamos la Venta o Cierre de Tienda 4 por los siguientes motivos:

1. **Facturación por debajo del promedio:** 23,6% vs. 25–26% de sus pares.
2. **Satisfacción del cliente:** Aunque tiene buena satisfacción 4.00, su facturación de 23,6% no justifica mantenerla frente a las otras 3 tiendas que generan más ingresos con calificaciones similares.
3. **Desempeño integral:** Puntuación de 62%, significativamente por debajo de otras tiendas.
4. **Oportunidad de costo:** Liberar capital, personal y recursos operativos para reinvertir.

### Beneficios esperados

| Beneficio | Impacto |
|-----------|---------|
| **Reducción de costos operativos** | Cierre de 1 punto de venta física |
| **Reubicación de personal** | Transferir vendedores a Tiendas 1–3 |
| **Concentración de inventario** | Optimizar stock en tiendas con mayor demanda |
| **Mejora de eficiencia** | Reducir complejidad operativa |
| **Potencial de crecimiento** | Reinvertir ahorros en canales digitales o tiendas de mayor rendimiento |

---

## 9. CONCLUSIÓN

AluraStore opera con un portafolio desequilibrado donde la Tienda 4 funciona por debajo de estándares. Su venta o cierre liberará recursos sin afectar facturación crítica, permitiendo enfoque en:

- Fortalecer canales digitales.
- Crecer en geografías de mayor potencial.

**La decisión de cerrar Tienda 4 es financieramente sólida y operativamente viable.**
